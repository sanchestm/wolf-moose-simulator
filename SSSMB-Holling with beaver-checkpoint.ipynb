```python
%pylab inline
from scipy.integrate import odeint
from numpy import *
import matplotlib
matplotlib.style.use("ggplot")
```

\begin{align*}
& WSat = 1+\sum_{j=S,I,B} h_ja_jN_j \\
& \frac{dN_S}{dt} = r_S(N_S + N_I)\left(1-\frac{N_S+N_I}{K_S}\right) - \frac{a_SN_S(P_S+P_I)}{WSat} -\beta N_SP_I \\
& \frac{dN_I}{dt} = \beta N_SP_I - \frac{a_IN_I(P_S+P_I)}{WSat} - \mu_M N_I \\
& \frac{dP_S}{dt} = \frac{e(P_S+P_I)}{WSat} \left[\sum_{j=S,I,B} a_j N_j \right] - \mu_p P_S -\frac{a_IN_IP_S}{WSat} \\
& \frac{dP_I}{dt} = \frac{a_IN_IP_S}{WSat} - \mu_p P_I \\
& \frac{dN_B}{dt} = r_BN_B\left(1-\frac{N_B}{K_B}\right) - \frac{a_BN_B(P_S+P_I)}{WSat} \\
\end{align*}

\begin{align*}
&\bullet N_j \Rightarrow \text{Prey j, j=I,B,S} \\
&\bullet P_j \Rightarrow \text{Predator of type j, j=I,S} \\
&\bullet h_j \Rightarrow \text{Time to handle prey j, j=I,S,B} \\
&\bullet a_j \Rightarrow \text{Rate to attack prey j, j=I,S,B} \\
&\bullet \beta \Rightarrow \text{Parasite Infection Rate per Population} \\
&\bullet r_j \Rightarrow \text{Generation Rate of preys j, j=S,B} \\
&\bullet \mu_j \Rightarrow \text{Death Rate for species j, j=M,P} \\
&\bullet K_j \Rightarrow \text{Carrying capacity of preys j, j=B,S} \\
&\bullet e = \text{Predator Efficiency}
\end{align*}

```python
def derivate(x,t): #remember to exclude preference for other stuff
    Ms, Mi, Ps, Pi, B = x
    wsat = 1./(1.+hm*pref/2*(As*Ms + Ai*Mi)+(1-pref)*hb*Ab*B)
    dMs = rm*(Ms+Mi)*(1.-(Ms+Mi)/km) - (pref/2)*As*Ms*(Ps+Pi)*wsat - bet*Ms*Pi
    dMi = bet*Ms*Pi - (pref/2)*Ai*Mi*(Ps+Pi)*wsat - mum*Mi
    dPs = ee*wsat*(pref/2*(As*Ms + Ai*Mi) + (1-pref)*Ab*B)*(Ps+Pi) - (mup+pref/2*Ai*Mi*wsat)*Ps
    dPi = pref/2*Ai*Mi*wsat*Ps - mup*Pi
    dB = rb*B*(1-B/kb) - (1-pref)*Ab*B*(Ps+Pi)*wsat
    return array([dMs, dMi, dPs, dPi, dB])
```

```python
rm=1.5
rb=3
Ai=0.09
As=0.04
Ab=0.09
hm=0.01
hb=0.03
bet=0.05
ee=0.05
mum=0.1
mup=0.1
km=60
kb=100
#inicial condition x_0=[NS_0,NI_0,PS_0,PI_0,B_0]
y0=[20,5,5,2,12]
t=arange(0,500,0.1)
```

```python
totalW = []
infectedmoose = []
preference = []
for pref in linspace(0, 1,70):
    #print("preference for moose:"+str(pref))
    preference += [pref]
    Y = odeint(derivate, y0, t)
    Y = Y.T
    #plot(t, [Y[0][i]+Y[1][i] for i in range(len(t))], label='Mtotal')
    #plot(t, [Y[2][i]+Y[3][i] for i in range(len(t))], label='Ptotal')
    #plot(t, Y[4], label='B')
    #legend(loc='best')
    #xlabel("time")
    #ylabel("Population size")
    #show()
    #plot(t, [Y[1][i]/(Y[0][i]+ Y[1][i]) for i in range(len(t))], label='infected moose')
    #plot(t, [Y[3][i]/(Y[2][i]+Y[3][i]) for i in range(len(t))], label='infected wolf')
    totalW += [Y[3][-1]+Y[2][-1]]
    infectedmoose += [Y[1][-1]/(Y[0][-1]+Y[1][-1])]
    #title("percentage of infected wildlife")
    #xlabel("time")
    #ylabel("percentage")
    #legend(loc='best')
    #show()

    
plot( preference,totalW, label = "wolf")
#plot( preference,infectedmoose, label = "moose")
xlabel("preference")
ylabel("wolfs population")
savefig("preference")
legend(loc = "best")
show()
```

```python
def a_funcao(a0,a,b,c,d,e,f,g,h,i,j,k,l):
    rm =a0
    rb=a
    Ai=b
    As=c
    Ab=d
    hm=e
    hb =f
    bet =g
    ee =h
    mum =i
    mup = j
    km =k
    kb =l
    t=arange(0,20,0.1)
    Y=odeint(derivate,y0,t)
    Y=Y.T
    plot(t, [Y[0][p]+Y[1][p] for p in range(len(t))], label='M')
    plot(t, [Y[2][p]+Y[3][p] for p in range(len(t))], label='P')

    plot(t, Y[4], label='B')
    legend(loc='best')
    xlabel("time")
    ylabel("Population size")
    show()
    plot(t, [Y[1][i]/(Y[0][i]+ Y[1][i]) for i in range(len(t))], label='infected moose')
    plot(t, [Y[3][i]/(Y[2][i]+Y[3][i]) for i in range(len(t))], label='infected wolf')
    xlim([0,10])
    title("percentage of infected wildlife")
    xlabel("time")
    ylabel("percentage")
    legend(loc='best')
    show()
```
